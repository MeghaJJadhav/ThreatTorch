```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path =r"C:\Users\dell\Desktop\NF-UQ-NIDS.csv"
df = pd.read_csv(path)
```

```python
column_to_drop = 'Dataset'
df = df.drop(column_to_drop, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# List of categorical columns
categorical_columns = ['PROTOCOL', 'IPV4_SRC_ADDR', 'L4_SRC_PORT', 'IPV4_DST_ADDR', 'L4_DST_PORT', 'L7_PROTO']

# Apply label encoding to each categorical column
label_encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python

X = df.drop(columns=['Label', 'Attack'])  
y = df['Label'] 
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold, cross_val_score
from xgboost import XGBClassifier
from sklearn.metrics import make_scorer, accuracy_score, classification_report

# Load your dataset
# Assuming your data is stored in a DataFrame called 'df'

# Initialize the Stratified K-Fold
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Define custom scorer for accuracy
scorer = make_scorer(accuracy_score)

# Initialize XGBoost classifier
model = XGBClassifier()

# Perform cross-validation with parallel processing
cv_results = cross_val_score(model, X, y, cv=kfold, scoring=scorer, n_jobs=-1)

# Print average accuracy
print(f"Average Accuracy: {np.mean(cv_results)}")

```

```python
import joblib

# Train the Random Forest model
model.fit(X, y)

# Save the trained model to a file
joblib.dump(model, 'xgboost.pkl')
```
