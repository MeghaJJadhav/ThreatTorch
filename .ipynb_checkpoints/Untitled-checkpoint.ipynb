```python
import pandas as pd
df = pd.read_excel('TTI_ALL_AGE_GRP.xlsx', sheet_name='allmale')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,8))
#sns.set(style='darkgrid', palette='colorblind', font='sans-serif', font_scale=1)

sns.boxplot(x='Site', y='tti', data=df)
sns.set(font='Times New Roman')

# Axis labels
plt.xlabel('Category')
plt.ylabel('Value')
plt.ylim(-10, 240)  # Add a bit of room below 0
plt.yticks(range(0, 250, 20))
plt.title('Boxplot of Value by Category')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.axes_style("ticks")
plt.figure(figsize=(10, 8))
sns.set(style='darkgrid', palette='colorblind', font='sans-serif', font_scale=1)

# Create boxplot
ax = sns.boxplot(x='Site', y='tti', data=df)

# Set font
sns.set(font='Times New Roman')

# Axis labels and formatting
plt.xlabel('Category')
plt.ylabel('Value')
plt.ylim(-10, 240)
plt.yticks(range(0, 250, 20))
plt.title('Boxplot of Value by Category')

# ➕ Add median values ABOVE the first gridline (e.g., y=20)
medians = df.groupby('Site')['tti'].median()
positions = range(len(medians))

for pos, (label, median) in zip(positions, medians.items()):
    plt.text(
        pos, 207, f'{median:.1f}',
        ha='center', va='bottom',
        fontsize=15, fontweight='bold',
        bbox=dict(facecolor='white', edgecolor='black')
    )

plt.show()
```
